```python
# importing neccessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading data and creating data frame
data_1  =pd.read_csv("C:\\Users\\vsman\\Downloads\\globalterrorismdb_0718dist.csv",encoding = 'latin1')
data_1.head()
```

```python
#Renaming columns in dataframe clearly
data_1.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City',
    'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound', 'natlty1_txt':'nationality'},inplace=True)
data_1
```

```python
data=data_1[['Year','Month','day','Group','Country','Region','State','City','latitude','longitude','summary','location','Attacktype','Targettype','Weapon','kill','Wound','nationality','motive','target1']]
data
```

```python
data.shape
```

```python
data.describe(include = 'all')
```

```python
data.isna().sum()
```

```python
data['Wound']=data['Wound'].fillna(0)
data['kill']=data['kill'].fillna(0)
```

```python
data['total damage to lives']=data['Wound']+data['kill']
data[['Wound','kill','total damage to lives']].head()
```

```python
# Lineplot for attacks over the years
import seaborn as sns
plt.figure(figsize = (50,10))
numberofyears =data['Year'].unique()
numberofkills =data['Year'].value_counts()
sns.lineplot(x= numberofyears, y= numberofkills)
plt.title('History of attacks over the years')
plt.xlabel('Years')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Finding out the most affected region
temp=data[['Region','total damage to lives']].groupby(['Region'],axis=0).sum().sort_values('total damage to lives',ascending=False).head(10)
plt.figure(figsize=(20,10))
sns.barplot(x=temp.index,y=temp.values.reshape(-1))
plt.title("Affected region",fontsize=20)
plt.xlabel('Region',fontsize=20)
plt.ylabel('Number of People Died',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
#finding out the most attacked country
attack=data.Country.value_counts()[:10]
plt.figure(figsize=(20,10))
sns.barplot(x=attack.index,y=attack.values)
plt.title('Most Attacked Country',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
# Finding the most attacked targettype
hel = data['Targettype'].value_counts().drop('Unknown')[:10]  
plt.pie(hel.values, labels=hel.index)
plt.show()
```

```python
# Most number of attacks done 
ty = data['Group'].value_counts()[:20]
plt.figure(figsize = (50,10))
sns.barplot(x=ty.index, y=ty.values,color ='red')
plt.xlabel('City name',fontsize = 40)
plt.ylabel('Kills',fontsize = 40)
plt.title('Groups involved in attack')
plt.xticks(rotation = 90)
plt.rcParams.update({'font.size': 40}) 
plt.show()
```

```python
# What type of attack is carried out the most
plt.rcParams['figure.figsize']=14,10
sns.countplot(y=data['Attacktype'], palette='inferno')
plt.show()
```

```python
#which is the most affected city
ty = data['City'].value_counts().drop('Unknown')[:15]
plt.figure(figsize = (50,10))
explode = (0.1,0,0.2,0.3,0.2)
plt.pie(ty.values, labels=ty.index)
plt.show()
```
